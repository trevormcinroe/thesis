```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
```

```python
files = os.listdir('../experiments/bounds/sac/upper_grasp/') 
files
```

```python
data = {}

for file in files:
    with open(f'../experiments/bounds/sac/upper_grasp/{file}', 'rb') as f:
        data[file] = pickle.load(f) 
```

```python
for d in data:
    print(len(data[d]))
```

```python
w = 200

smoothed_data = {}

for file in data.keys():
    b_idx = 0
    e_idx = 0 + w
    
    smoothed_data[file] = []
    
    while e_idx < len(data[file]):
        smoothed_data[file].append(np.mean(data[file][b_idx:e_idx]))
        b_idx += 1
        e_idx += 1
```

```python
lst = list(smoothed_data.keys())
lst
```

```python
averages = []
stds = []

for i in range(len(smoothed_data[lst[0]])):
    averages.append(np.nanmean([x[i] for x in smoothed_data.values()]))
    stds.append(np.std([x[i] for x in smoothed_data.values()]))
```

```python
plt.plot(averages)
plt.fill_between(range(len(smoothed_data[lst[0]])),
                 [x if x <= 1.0 else 1.0 for x in np.array(averages) + np.array(stds)],
                 np.array(averages) - np.array(stds), alpha=0.25)

plt.legend(['Proprioceptive'])

```
